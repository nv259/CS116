### 1. Tải dữ liệu

```python
# Import những thư viện cần thiết
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from PIL import Image
```

```python
# Tải dữ liệu hoa Iris
iris = load_iris()
ans_path = "data"
df = pd.DataFrame(
    data=np.c_[iris["data"], iris["target"]],
    columns=iris["feature_names"] + ["target"]
)
# Xem qua dữ liệu
df.head()
```

### 2. Bài tập vẽ biểu đồ

**Bài tập**: Hãy sử dụng thư viện sns để tạo một histogram đối với feature ***sepal length (cm)***
với các tham số như sau:

- ***figsize***: (8, 6)
- ***title***: "Histogram of Sepal Length"
- ***xlabel***: "Sepal Length (cm)"
- ***ylable***: "Frequency"

Lưu ý: Không thêm bất kỳ tùy chỉnh nào khác cho biểu đồ để thuận tiện cho việc chấm tự động.

```python
# YOUR SOLUTION HERE
from io import BytesIO
### BEGIN SOLUTION
sns.set(rc={"figure.figsize": (8, 6)})
sns.histplot(data=df["sepal length (cm)"]).set(title="Histogram of Sepal Length",
                                              ylabel="Frequency",
                                              xlabel="Sepal Length (cm)")
### END SOLUTION

# Không chỉnh những dòng code này
buf = BytesIO()
plt.savefig(buf, format='png')
buf.seek(0)
image1 = Image.open(buf)
plt.show()
```

**Bài tập**: Hãy sử dụng thư viện sns để tạo một boxplot đối với feature ***sepal length (cm)***
với các tham số như sau:

- ***figsize***: (8, 6)
- ***title***: "Box Plot of Petal Width"
- ***ylable***: "Sepal Length (cm)"

Lưu ý: Không thêm bất kỳ tùy chỉnh nào khác cho biểu đồ để thuận tiện cho việc chấm tự động.

```python
# YOUR SOLUTION HERE

### BEGIN SOLUTION
sns.set(rc={"figure.figsize": (8, 6)})
sns.boxplot(data=df["sepal length (cm)"]).set(title="Box Plot of Petal Width",
                                              ylabel="Sepal Length (cm)")
### END SOLUTION

# Không chỉnh những dòng code này
buf = BytesIO()
plt.savefig(buf, format='png')
buf.seek(0)
image1 = Image.open(buf)
plt.show()
```

**Bài tập**: Hãy sử dụng thư viện seaborn để tạo một kernel density plot (kde) đối với feature ***sepal width (cm)*** với các tham số như sau:

- ***figsize***: (8, 6)
- ***title***: "Kernel Density Estimate of Sepal Width"
- ***xlabel***: "Sepal Width (cm)"
- ***ylabel***: "Density"

Lưu ý: Không thêm bất kỳ tùy chỉnh nào khác cho biểu đồ để thuận tiện cho việc chấm tự động.

```python
# YOUR SOLUTION HERE

### BEGIN SOLUTION
sns.set(rc={"figure.figsize": (8, 6)})
sns.kdeplot(data=df['sepal width (cm)']).set(title="Kernel Density Estimate of Sepal Width",
                                             xlabel="Sepal Width (cm)",
                                             ylabel="Density")
### END SOLUTION

# Không chỉnh những dòng code này
buf = BytesIO()
plt.savefig(buf, format='png')
buf.seek(0)

image1 = Image.open(buf)
plt.show()
```

**Bài tập**: Hãy sử dụng thư viện seaborn để tạo count plot đối với feature ***target***
với các tham số như sau:
- ***figsize***: (8, 6)
- ***title***: "Count Plot of Target Classes"
- ***xlabel***: "Target Class"
- ***ylabel***: "Count"
Lưu ý: Không thêm bất kỳ tùy chỉnh nào khác cho biểu đồ để thuận tiện cho việc chấm tự động.

**Bài tập**: Hãy sử dụng thư viện seaborn để tạo violin plot cho feature ***petal length*** cho mỗi feature ***target*** với các tham số như sau:

- ***ylabel***: "Petal Length (cm)"
